```python
def get_adjacent(el):
    a = el[0]
    b = el[1]
    ls_adj = [(x,y) for x in range(a-1,a+2) for y in range(b-1,b+2) if x in range(0,10) if y in range(0,10)]
    ls_adj.remove(el)
    
    return ls_adj
```

```python
df = df_input.copy()
result = 0
step = 0
result_step = 0

while result_step != df.shape[0]*df.shape[1]:
    step += 1
    df = df.applymap(increase)

    lst1 = [(x, df.columns[y]) for x, y in zip(*np.where(df.values > 9))]
    lst2 = [(x, df.columns[y]) for x, y in zip(*np.where(df.values < 999))]
    ls_flash = [value for value in lst1 if value in lst2]

    while ls_flash:

        for el in ls_flash:
            ls_adj = get_adjacent(el)
            for xy in get_adjacent(el):
                df.loc[xy] += 1

            df.loc[el] = 999

        lst1 = [(x, df.columns[y]) for x, y in zip(*np.where(df.values > 9))]
        lst2 = [(x, df.columns[y]) for x, y in zip(*np.where(df.values < 999))]
        ls_flash = [value for value in lst1 if value in lst2]

    ls_flash_step = [(x, df.columns[y]) for x, y in zip(*np.where(df.values >= 999))]
    result_step = len(ls_flash_step)
    result += result_step
    #print(result_step)
    df = df.applymap(set_zero)

step
```
