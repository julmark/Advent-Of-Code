# Day 13: Transparent Origami

```python
df_input = pd.read_csv('20211213_aoc_input_1.csv',header=None,dtype=str)
```

```python
#df_coord = df_input.loc[:len(df_input)-3,:]
df_coord = df_input.loc[:len(df_input)-13,:]
df_coord.columns = ['c','r']
df_coord = df_coord.astype(int)
#df_fold = df_input[0].loc[len(df_input)-2:]
df_fold = df_input[0].loc[len(df_input)-12:]
```

```python
shape_c = df_coord.c.max()
shape_r = df_coord.r.max()
```

```python
df = pd.DataFrame([['.' for x in range(shape_c+1)] for y in range(shape_r+1)],columns=[i for i in range(shape_c+1)])
```

```python
for r in df_coord.index:
    df.loc[df_coord.loc[r,'r'],df_coord.loc[r,'c']] = '#'
```

```python
# fold along y=7
def fold_along_y(df,df_coord,y):
    df_coord.r.loc[df_coord.r > y] = df_coord.r.loc[df_coord.r > y].apply(lambda x: x - 2*abs(x - y))
    for r in df_coord.index:
        df.loc[df_coord.loc[r,'r'],df_coord.loc[r,'c']] = '#'

    return df.loc[:y-1,:], df_coord
```

```python
# fold along x=655
def fold_along_x(df,df_coord,x_line):
    df_coord.c.loc[df_coord.c > x_line] = df_coord.c.loc[df_coord.c > x_line].apply(lambda x: x - 2*abs(x - x_line))
    for r in df_coord.index:
        df.loc[df_coord.loc[r,'r'],df_coord.loc[r,'c']] = '#'

    return df.loc[:,:x_line-1], df_coord
```

```python
df, df_coord = fold_along_x(df,df_coord,655)
df, df_coord = fold_along_y(df,df_coord,447)
df, df_coord = fold_along_x(df,df_coord,327)
df, df_coord = fold_along_y(df,df_coord,223)
df, df_coord = fold_along_x(df,df_coord,163)
df, df_coord = fold_along_y(df,df_coord,111)
df, df_coord = fold_along_x(df,df_coord,81)
df, df_coord = fold_along_y(df,df_coord,55)
df, df_coord = fold_along_x(df,df_coord,40)
df, df_coord = fold_along_y(df,df_coord,27)
df, df_coord = fold_along_y(df,df_coord,13)
df, df_coord = fold_along_y(df,df_coord,6)
```
